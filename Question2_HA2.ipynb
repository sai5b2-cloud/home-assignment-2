```python
#Task1
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

def sobel_edge_detection():
    # Use a built-in sample image (grayscale)
    image = data.camera()  # This is a built-in grayscale image

    # Define Sobel filters
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]])

    sobel_y = np.array([[-1, -2, -1],
                        [0, 0, 0],
                        [1, 2, 1]])

    # Apply filters
    edges_x = cv2.filter2D(image, -1, sobel_x)
    edges_y = cv2.filter2D(image, -1, sobel_y)

    # Display results
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(edges_x, cmap='gray')
    plt.title('Sobel X')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(edges_y, cmap='gray')
    plt.title('Sobel Y')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

sobel_edge_detection()
```

```python
#Task2
import tensorflow as tf
import numpy as np

def pooling_operations():
    # Create a random 4x4 matrix
    np.random.seed(42)  # for reproducibility
    input_matrix = np.random.rand(1, 4, 4, 1)  # batch=1, height=4, width=4, channels=1

    print("Original Matrix:")
    print(np.squeeze(input_matrix))  # Remove single-dimensional entries

    # Max Pooling
    max_pool = tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=2)
    max_pooled = max_pool(input_matrix)

    print("\nMax Pooled Matrix:")
    print(np.squeeze(max_pooled.numpy()))

    # Average Pooling
    avg_pool = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=2)
    avg_pooled = avg_pool(input_matrix)

    print("\nAverage Pooled Matrix:")
    print(np.squeeze(avg_pooled.numpy()))

pooling_operations()
```
